```python
# Created for anomaly detection 
plt.figure(figsize=(12, 5))

# Weight Boxplot
plt.subplot(1, 2, 1)
sns.boxplot(data=df, y='Kilo')
plt.title('Weight Boxplot')
plt.ylabel('Kilo')

# Height Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(data=df, y='Boy')
plt.title('Height Boxplot')
plt.ylabel('Boy')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating Age Group
df['Ya'] = (pd.Timestamp.now() - df['Dogum_Tarihi']).dt.days // 365
df['Ya Grubu'] = pd.cut(df['Ya'], bins=[0, 18, 35, 50, 65, 100], 
                         labels=['0-18', '19-35', '36-50', '51-65', '66+'])

# Side effect rates
yan_etki_oranlar = df.groupby('Ya Grubu')['Yan_Etki'].value_counts(normalize=True).unstack().fillna(0)

yan_etki_oranlar.plot(kind='bar', stacked=True)
plt.title('Ya Gruplarna Gre Yan Etki Oranlar')
plt.xlabel('Ya Grubu')
plt.ylabel('Oran')
plt.legend(title='Yan Etki')
plt.show()
```
