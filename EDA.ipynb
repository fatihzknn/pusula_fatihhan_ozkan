```python


#creating Age Group
df['Yaş'] = (pd.Timestamp.now() - df['Dogum_Tarihi']).dt.days // 365
df['Yaş Grubu'] = pd.cut(df['Yaş'], bins=[0, 18, 35, 50, 65, 100], 
                         labels=['0-18', '19-35', '36-50', '51-65', '66+'])

#side effect rates
yan_etki_oranları = df.groupby('Yaş Grubu')['Yan_Etki'].value_counts(normalize=True).unstack().fillna(0)

#most 5
en_yuksek_yan_etkiler = yan_etki_oranları.sum().nlargest(5).index


filtered_yan_etki_oranları = yan_etki_oranları[en_yuksek_yan_etkiler]


filtered_yan_etki_oranları.plot(kind='bar', figsize=(10, 6))
plt.title('Side Effect Rates by Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Rate')
plt.legend(title='Yan Etki', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

```python

ilac_kullanimi = df.groupby(['Il', 'Ilac_Adi']).size().reset_index(name='Kullanim_Sayisi')

en_cok_kullanilan_ilac = ilac_kullanimi.loc[ilac_kullanimi.groupby('Il')['Kullanim_Sayisi'].idxmax()]

plt.figure(figsize=(12,8))
sns.barplot(x='Kullanim_Sayisi', y='Il', hue='Ilac_Adi', data=en_cok_kullanilan_ilac, palette='Set2')
plt.title('Most Used Drugs in Cities')
plt.xlabel('Count of using')
plt.ylabel('Cities')
plt.legend(title='Drug Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
pivot_table = en_cok_kullanilan_ilac.pivot(index='Il', columns='Ilac_Adi', values='Kullanim_Sayisi').fillna(0)

plt.figure(figsize=(12,8))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt='.0f', linewidths=.5)
plt.title('Most Used Drugs in Cities (heatmap)')
plt.xlabel('Drug Name')
plt.ylabel('Cities')
plt.tight_layout()
plt.show()
```

The Most Common Allergies For Men and Women

```python
male_allergies = df[df['Cinsiyet'] == 'Male'].groupby('Alerjilerim').size().reset_index(name='Alerji Sayısı')
female_allergies = df[df['Cinsiyet'] == 'Female'].groupby('Alerjilerim').size().reset_index(name='Alerji Sayısı')

top5_male_allergies = male_allergies.sort_values(by='Alerji Sayısı', ascending=False).head(5)
top5_female_allergies = female_allergies.sort_values(by='Alerji Sayısı', ascending=False).head(5)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

sns.barplot(x='Alerjilerim', y='Alerji Sayısı', data=top5_male_allergies, ax=axes[0], palette='Set1')
axes[0].set_title('The Most Common Allergies Men Have')
axes[0].set_xlabel('Allergies')
axes[0].set_ylabel('Count of Allergie')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')

sns.barplot(x='Alerjilerim', y='Alerji Sayısı', data=top5_female_allergies, ax=axes[1], palette='Set2')
axes[1].set_title('The Most Common Allergies Women Have')
axes[1].set_xlabel('Allergies')
axes[1].set_ylabel('Count of Allergie')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

Chronic Diseases Most Suffering from Men and Women

```python
male_diseases = df[df['Cinsiyet'] == 'Male'].groupby('Kronik Hastaliklarim').size().reset_index(name='Hastalık Sayısı')
female_diseases = df[df['Cinsiyet'] == 'Female'].groupby('Kronik Hastaliklarim').size().reset_index(name='Hastalık Sayısı')

top5_male_diseases = male_diseases.sort_values(by='Hastalık Sayısı', ascending=False).head(5)
top5_female_diseases = female_diseases.sort_values(by='Hastalık Sayısı', ascending=False).head(5)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

sns.barplot(x='Kronik Hastaliklarim', y='Hastalık Sayısı', data=top5_male_diseases, ax=axes[0], palette='Set1')
axes[0].set_title('Chronic Diseases Most Suffering from Men')
axes[0].set_xlabel('Chronic Diseases')
axes[0].set_ylabel('Count of Diseases')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')

sns.barplot(x='Kronik Hastaliklarim', y='Hastalık Sayısı', data=top5_female_diseases, ax=axes[1], palette='Set2')
axes[1].set_title('Chronic Diseases Most Suffering from Women')
axes[1].set_xlabel('Chronic Diseases')
axes[1].set_ylabel('Count of Diseases')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

Distribution of diseases by cities

```python
top5_diseases_by_il = df.groupby(['Il', 'Kronik Hastaliklarim']).size().reset_index(name='Hastalık Sayısı')
top5_diseases_by_il = top5_diseases_by_il.sort_values(by=['Il', 'Hastalık Sayısı'], ascending=False)
top5_diseases_by_il = top5_diseases_by_il.groupby('Il').head(5)

unique_ills = top5_diseases_by_il['Il'].unique()
num_ills = len(unique_ills)
fig, axes = plt.subplots(nrows=(num_ills // 3) + 1, ncols=3, figsize=(15, num_ills * 2))

for i, il in enumerate(unique_ills):
    ax = axes[i // 3, i % 3]
    sns.barplot(x='Kronik Hastaliklarim', y='Hastalık Sayısı', data=top5_diseases_by_il[top5_diseases_by_il['Il'] == il], ax=ax, palette='Set2')
    ax.set_title(f'{il} Distrubition of chronic diseases')
    ax.set_xlabel('chronic diseases')
    ax.set_ylabel('number of diseases')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

Most 5 Side Effect by Blood Groups

```python
df_filtered = df.dropna(subset=['Kan Grubu'])

kan_gruplari = df_filtered['Kan Grubu'].unique()

nrows = (len(kan_gruplari) + 2) // 3  
ncols = 3  

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 10))

axes = axes.flatten()  
for i in range(nrows * ncols):
    if i >= len(kan_gruplari):
        fig.delaxes(axes[i]) 
for i, kan_grubu in enumerate(kan_gruplari):
    ax = axes[i]
    
    top5_side_effects = df_filtered[df_filtered['Kan Grubu'] == kan_grubu]['Yan_Etki'].value_counts().nlargest(5)
    
    sns.barplot(x=top5_side_effects.index, y=top5_side_effects.values, 
                ax=ax, palette="Set3")
    
    ax.set_title(f'{kan_grubu} Kan Grubu')
    ax.set_xlabel('Side Effects')
    ax.set_ylabel('Count of side effects')
    ax.tick_params(axis='x', rotation=45) 

plt.tight_layout()
plt.show()
```
